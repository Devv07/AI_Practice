```python
import requests
import pandas as pd

url="https://jsonplaceholder.typicode.com/users";

response=requests.get(url)
print(response.status_code)

data=response.json()
df=pd.DataFrame(data)
df.to_csv("user_data.csv",index=False)

df.head()
```
