```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

# Step 1: Sample string data
texts = [
    "I love this product",
    "This is the best purchase I made",
    "Worst experience ever",
    "I hate this item",
    "Absolutely fantastic",
    "Terrible and disappointing"
]

# Step 2: Labels (target categories)
labels = ["Positive", "Positive", "Negative", "Negative", "Positive", "Negative"]

# Step 3: Convert text to numeric features using CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts).toarray()  # Convert to NumPy array
y = np.array(labels)  # Convert target to NumPy array

# Step 4: Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Step 5: Train a model
model = MultinomialNB()
model.fit(X_train, y_train)

# Step 6: Predict on new data
new_review = ["I really love it"]
new_vector = vectorizer.transform(new_review).toarray()

prediction = model.predict(new_vector)
print("Predicted label:", prediction[0])  # Output: Positive
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Step 1: Input text
text = "hello world this is ai powered text generation"

# Step 2: Create character mapping
chars = sorted(list(set(text)))
char_to_idx = {c: i for i, c in enumerate(chars)}
idx_to_char = {i: c for c, i in char_to_idx.items()}

# Step 3: Prepare sequences
seq_length = 5
sequences = []
next_chars = []

for i in range(len(text) - seq_length):
    seq = text[i:i + seq_length]
    next_char = text[i + seq_length]
    sequences.append([char_to_idx[c] for c in seq])
    next_chars.append(char_to_idx[next_char])

X = np.array(sequences)
y = to_categorical(next_chars, num_classes=len(chars))

# Step 4: Define model
model = Sequential()
model.add(LSTM(128, input_shape=(X.shape[1], 1)))
model.add(Dense(len(chars), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam')
X = np.reshape(X, (X.shape[0], X.shape[1], 1)) / len(chars)

# Step 5: Train
model.fit(X, y, epochs=100, verbose=0)

# Step 6: Generate new text
def generate_text(seed, length=50):
    for _ in range(length):
        input_seq = [char_to_idx.get(c, 0) for c in seed[-seq_length:]]
        input_seq = np.array(input_seq).reshape(1, seq_length, 1) / len(chars)
        predicted = model.predict(input_seq, verbose=0)
        next_idx = np.argmax(predicted)
        next_char = idx_to_char[next_idx]
        seed += next_char
    return seed

print("Generated text:")
print(generate_text("Hello"))
```

```python
import random

# Simulating a lottery draw for an IPO
# Total applicants: 50,000
# Available winners: 10,000 (10 units per person)

total_applicants = 50000
available_winners = 10000

# Generate mock BOIDs (13-digit numbers)
applicants = [f"{random.randint(1000000000000, 9999999999999)}" for _ in range(total_applicants)]

# Randomly select winners
winners = set(random.sample(applicants, available_winners))

# Show a small sample of applicants and which ones were selected
sample_check = random.sample(applicants, 10)
sample_results = {boid: ("✅ Allocated" if boid in winners else "❌ Not Allocated") for boid in sample_check}

# Count how many in the sample got allocated
allocated_count = sum(1 for result in sample_results.values() if "Allocated" in result)

sample_results, allocated_count, available_winners, total_applicants
```
